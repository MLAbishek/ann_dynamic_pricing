```python
import tensorflow as tf
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df=pd.read_csv(r"D:\programming\coding\dataset\dynamic_pricing.csv")
df.head()
```

```python
np.unique(df['Location_Category'])
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print(x.shape)
y=y.reshape(-1,1)
print(y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
x[:,2]=lr.fit_transform(x[:,2])
x[:,3]=lr.fit_transform(x[:,3])
x[:,6]=lr.fit_transform(x[:,6])
x[:,7]=lr.fit_transform(x[:,7])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
y_train=sc.fit_transform(y_train)
y_test=sc.transform(y_test)
```

```python
ann=tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=9,activation='relu'))
ann.add(tf.keras.layers.Dense(units=20,activation='relu'))
ann.add(tf.keras.layers.Dense(units=1,activation='linear'))
ann.compile(optimizer='adam',loss='mse',metrics=['mse'])
ann.fit(x_train,y_train,epochs=250,batch_size=32)
```

```python
y_pred=sc.inverse_transform(ann.predict(x_test))
print(y_pred)
```
